```python
import pandas as pd
from time import time
import numpy as np
from sklearn.preprocessing import minmax_scale, MinMaxScaler
from matplotlib import pyplot
import seaborn as sns
sns.set_style('ticks')

from keras.models import Sequential, Model
from keras.layers import Flatten, Dense, Conv1D, Input, Reshape, Dropout, LeakyReLU
from keras.optimizers import Adam, RMSprop, Nadam
from keras.utils import plot_model
from keras.callbacks import TensorBoard, ModelCheckpoint
from keras.layers.merge import concatenate
```

```python
raw_df = pd.read_csv('data/history/ADABTC.csv')
raw_df.head()
```

```python
ndf = raw_df.filter(items=['Open', 'High', 'Low', 'Close', 'Volume', 'Number Trades', 'Quote Asset Volume', 'Taker Base Asset Volume', 'Taker Quote Asset Volume'])
ndf = ndf.set_index(pd.to_datetime(raw_df['Open Time'],unit='ms'))
ndf.dropna(inplace=True)
ndf.head()
```

```python
# specify columns to plot
cols = ndf.columns
i = 1
# plot each column
pyplot.figure(figsize=(30, 10), dpi=400)
pyplot.subplots_adjust(hspace = 1)
for col in cols:
    pyplot.subplot(len(cols), 1, i)
    pyplot.plot(ndf[col])
    pyplot.title(col, y=1, loc='left')
    i += 1
pyplot.show()
```

```python
open_scaler = MinMaxScaler()
open_scaler.fit(X=ndf['Open'].values.reshape(-1, 1))
andf = ndf.apply(lambda c: minmax_scale(X=c, feature_range=(0, 1)))
andf.head()
```

```python
def frame_to_features(df, time_in, time_out):
    f = []
    n_features = len(df.columns)
    for i in range(time_in, len(andf)):
        # reshape input to be 3D [samples, timesteps, features]
        f.append([
            np.array(df.iloc[i - time_in:i].values),
            np.array(df['Open'].iloc[i - time_out + 2:i + 2].values)
        ])
    return f
```

```python
history_length = 10
prediction_length = 1
fs = frame_to_features(andf, history_length, prediction_length)
```

```python
# split into train and test sets
train = [x[0] for x in fs][:-1]
test = [x[1] for x in fs][:-1]
# split into input and outputs
train_X, test_X = np.array(train[:int(len(train)*0.7)]), np.array(train[int(len(train)*0.7):])
train_y = np.array(test[:int(len(test)*0.7)])
test_y  = np.array(test[int(len(test)*0.7):])
```

```python
kernel_length = 5

history_cnn_input = Input(shape=(train_X.shape[1], train_X.shape[2]))
# history_cnn_out = Reshape((1, self.state_size))(history_cnn_input)
history_cnn_out = Conv1D(filters=64, kernel_size=kernel_length, padding='same')(history_cnn_input)
history_cnn_out = LeakyReLU(alpha=0.1)(history_cnn_out)
history_cnn_out = Conv1D(filters=48, kernel_size=kernel_length, padding='same')(history_cnn_out)
history_cnn_out = LeakyReLU(alpha=0.1)(history_cnn_out)
history_cnn_out = Conv1D(filters=32, kernel_size=kernel_length, padding='same')(history_cnn_out)
history_cnn_out = LeakyReLU(alpha=0.1)(history_cnn_out)
history_cnn_out = Flatten()(history_cnn_out)
history_cnn_out = Dense(max(int(history_length * 0.5), 10))(history_cnn_out)
history_cnn_out = LeakyReLU(alpha=0.1)(history_cnn_out)
# history_cnn_out = Dropout(0.1)(history_cnn_out)
history_cnn_out = Dense(max(int(history_length * 0.2), 10))(history_cnn_out)
history_cnn_out = LeakyReLU(alpha=0.1)(history_cnn_out)
# history_cnn_out = Dense(10)(history_cnn_out)
# history_cnn_out = LeakyReLU(alpha=0.1)(history_cnn_out)
history_cnn_out = Dense(prediction_length, activation='linear')(history_cnn_out)
brain = Model(input=history_cnn_input, output=history_cnn_out)
brain.compile(optimizer=Nadam(lr=0.001), metrics=['mae', 'mse', 'accuracy'], loss='mae')
plot_model(brain, show_shapes=True, to_file='brain.png')
```

```python
# fit network
history = brain.fit(train_X, train_y, epochs=10, batch_size=32, validation_data=(test_X, test_y), verbose=1, shuffle=False)
# plot history
pyplot.plot(history.history['loss'], label='train')
pyplot.plot(history.history['val_loss'], label='test')
pyplot.legend()
pyplot.show()
```

```python
# # make a prediction
open_X = test_X
open_y = [np.nan]*len(open_X)
i = 50
yhat = brain.predict(open_X[i].reshape(1, history_length, -1))
invyhat = open_scaler.inverse_transform(yhat)
invyhat = [invyhat]*5000
for e, v in enumerate(invyhat):
    open_y[i+e] = v

pyplot.figure(figsize=(30, 10), dpi=400)
pyplot.plot(ndf['Open'].values, label='real')
pyplot.plot(open_y, label='predicted')
pyplot.legend()
pyplot.show()
```
